```python
import random
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import io
import requests
import math
from scipy.stats import norm
from random import randint
from scipy.stats import binom
import statistics
import pandas as pd
```

```python
indoor_data = pd.read_csv('BasementFullDataSet.csv')
```

```python
indoor_data
```

```python
#Graphing Temperature Data
plt.plot(indoor_data["Temperatures(Celsius)"])
plt.ylabel('Temperature (Celsius)')            
plt.xlabel('Device Time')
plt.title("Temperatures (Celsius) Vs. Device Time")    
plt.show()

plt.hist(indoor_data["Temperatures(Celsius)"], bins = 40)
plt.xlabel('Temperature (Celsius)')            
plt.ylabel('Frequency')
plt.title("Frequency Vs. Temperatures (Celsius)")    

def indoor_Gaus_distribution(x , mean , sd):
    Px = (288*(3.14*sd) * np.exp(-0.5*((x-mean)/sd)**2))
    return Px

temp_mu = np.mean(indoor_data["Temperatures(Celsius)"])
temp_std = np.std(indoor_data["Temperatures(Celsius)"])

x = np.linspace(np.min(indoor_data["Temperatures(Celsius)"]), np.max(indoor_data["Temperatures(Celsius)"]), 150)

Temp_gauss = indoor_Gaus_distribution(x,temp_mu,temp_std)

plt.plot(x, Temp_gauss, color = 'red')
plt.xlabel('Temperatures (celsius)')
plt.ylabel('Frequency')
plt.show()
```

```python

plt.plot(indoor_data["Pressure(Hectopascals)"])
plt.ylabel('Pressure (HectoPascals)')            
plt.xlabel('Device Time')
plt.title("Pressure (HectoPascals) Vs. Device Time")                             
plt.show()

plt.hist(indoor_data["Pressure(Hectopascals)"], bins = 25)
plt.xlabel('Pressure (HectoPascals)')            
plt.ylabel('Frequency')
plt.title("Frequency Vs. Pressure (HectoPascals)")                             

press_mu = np.mean(indoor_data["Pressure(Hectopascals)"])
press_std = np.std(indoor_data["Pressure(Hectopascals)"])

x = np.linspace(np.min(indoor_data["Pressure(Hectopascals)"]), np.max(indoor_data["Pressure(Hectopascals)"]), 150)

press_gauss = indoor_Gaus_distribution(x,press_mu,press_std)

plt.plot(x, press_gauss, color = 'red')
plt.show()
```
